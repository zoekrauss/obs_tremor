```python
import datetime
from obspy.clients.fdsn.client import Client
import obspy
import matplotlib.pyplot as plt
from netCDF4 import Dataset, num2date
import matplotlib
import numpy as np
matplotlib.rcParams['font.family']=['Arial']
from dateutil.relativedelta import *
import warnings
warnings.filterwarnings('ignore')
import requests
import pickle
matplotlib.rcParams.update({'font.size': 22})
import scipy
```

## Read wind data

```python
with open('era5_hysb1_windspeeds.pickle', 'rb') as handle:
    sb_ws_times,sb_ws = pickle.load(handle)
```

```python
with open('era5_hys14_windspeeds.pickle', 'rb') as handle:
    hr_ws_times,hr_ws = pickle.load(handle)
```

## Read bottom current data

```python
with open('hysb1_bottomcurrents.pickle', 'rb') as handle:
    sb_current_times,sb_current_max,sb_current_med=pickle.load(handle)
```

```python
with open('hys14_bottomcurrents.pickle', 'rb') as handle:
    hr_current_times,hr_current_max,hr_current_med=pickle.load(handle)
```

## Plot!

```python
%matplotlib widget

fig, ax = plt.subplots(2,1,figsize=(15,10),sharex=True)

# ax[0].plot([t.datetime for t in hr_current_times],hr_current_max,linewidth=1,label='Hourly Maximum')
ax[0].plot([t.datetime for t in hr_current_times],hr_current_med,linewidth=0.3,label='HYS14',color='royalblue')
ax[0].plot([t.datetime for t in sb_current_times],sb_current_med,linewidth=0.3,label='HYSB1',color='darkorange')
ax[0].set_ylabel('Bottom currents (m/s)')
ax[0].legend()
ax[0].set_title('Hourly bottom currents')

ax[1].plot(hr_ws_times,hr_ws,linewidth=0.3,label='HYS14',color='royalblue')
ax[1].plot(sb_ws_times,sb_ws,linewidth=0.3,label='HYSB1',color='darkorange')
ax[1].set_ylabel('Wind speeds (m/s)')
ax[1].legend()
ax[1].set_title('Hourly wind speeds')

ax[0].set_xlim([datetime.datetime(2015,1,1),datetime.datetime(2024,1,1)])

plt.savefig('environmental_data.eps', bbox_inches='tight',dpi=400)
```
